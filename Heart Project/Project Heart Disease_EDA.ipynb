```python
### rename the col names

heart=heart_disease.rename({"HeartDisease":"hd",'BMI':'bmi','Smoking':'smoke','AlcoholDrinking':'alcohol','Stroke':'str',
                                    'PhysicalHealth':'phy_health','MentalHealth':'ment_health','DiffWalking':'diff_walk','Sex':'sex',
                                    'AgeCategory':'age','PhysicalActivity':'phy_activ','GenHealth':'gen_health','Race':'r',"Diabetic":'d',
                                    'SleepTime':'st',"Asthma":'asth','KidneyDisease':'kd','SkinCancer':'sc'},axis=1)
heart
```

```python
### get unique value & their count

print(heart.age.nunique())
print(heart.age.unique())           ## Age
print(heart.age.value_counts())
sns.countplot(heart.age)             ## Visualize the count
```

# Visualization

```python
### LINE PLOT
sns.relplot(x="bmi", y="st", kind="line", data=heart)     ## bmi vs st
```

```python
sns.relplot(x="phy_health", y="ment_health", hue="hd", kind="line", data=heart)     ## phy_health vs ment_health vs hd
```

```python
### BOX PLOT

plt.figure(figsize=(8,6))                 #BMI
plt.boxplot(heart.bmi)
plt.ylabel("freq")
plt.xlabel("BMI")
plt.title("BOX Plot for BMI")
```

```python
plt.figure(figsize=(8,6))                     # Phy_Health
plt.boxplot(heart.phy_health)
plt.ylabel("freq")
plt.xlabel("Phy_Health")
plt.title("BOX Plot for Phy_Health")
```

```python
plt.figure(figsize=(8,6))                   # Ment_Health
plt.boxplot(heart.ment_health)
plt.ylabel("freq")
plt.xlabel("Ment_Health")
plt.title("BOX Plot for Ment_Health")
```

```python
plt.figure(figsize=(8,6))                 # st
plt.boxplot(heart.st)
plt.ylabel("freq")
plt.xlabel("st")
plt.title("BOX Plot for st")
```

```python
### create Histogram

heart.hist(figsize=(12, 8))
```

```python
plt.figure(figsize=(15,12))      ### conditioning on other variables
sns.displot(heart,bins=10)
plt.xlabel("Heart dataset")
plt.ylabel(" count")
plt.title("Dist plot")
```

```python
plt.figure(figsize=(15,12))
sns.displot(heart, bins=10, element="step")
plt.xlabel("Heart dataset")
plt.ylabel(" count")
plt.title("Dist plot")
```

```python
plt.figure(figsize=(15,12))
sns.displot(heart, kind="kde", bw_adjust= 2)   ## Density plot
plt.xlabel("Heart dataset")
plt.ylabel("Density")
plt.title("Density plot")
```

```python
### Scatter plot

plt.scatter(heart['phy_health'],heart['ment_health'])         ## Phy_Health vs Ment_Health
plt.xlabel("Phy_Health")
plt.ylabel("Ment_Health")
plt.title("Scatter plot of Phy_Health vs Ment_Health")
```
